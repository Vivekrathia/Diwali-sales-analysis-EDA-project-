```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
#checking for null values

pd.isnull(df).sum()
```

```python
df.shape
```

```python
#deleting the null values from Amount as it has 12 Null values

df.dropna(inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
# changing datatype from float64 to int64
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.info()
```

```python
# to rename any column in dataframe
df.rename(columns = {'Marital_Status': 'Married/Not-Married'})
```

```python
df.describe()
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

EXPLORATORY DATA ANALYSIS

we will perform analysis based on Gender, Age and State and Married Status

```python
graph0 = sns.countplot(x = 'Gender', data = df)
for bars in graph0.containers:
    graph0.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending =False)
```

```python
graph2 = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending =False)
sns.barplot(x = 'Gender', y = 'Amount', data = graph2)
```

from above graph we can see that most of the buyers are Female

```python
df.columns
```

```python
graph3 = sns.countplot(data=df, x = 'Age Group', hue = 'Gender')
for bars in graph3.containers:
    graph3.bar_label(bars)
```

```python
#Amount vs age group
sales_vs_age = df.groupby(['Age Group'], as_index =False)['Amount'].sum().sort_values(by='Amount', ascending =False)

sns.barplot(data= sales_vs_age, x = 'Age Group', y='Amount')
```

FROM ABOVE 2 GRAPHS WE CAN SEE THAT, PEOPLE WHO SPENT THE MOST ARE FROM AGE-GROUP 26-35 AND ARE FEMALES

```python
df.columns
```

```python
state_vs_amount = df.groupby(['State'], as_index =False)['Amount'].sum().sort_values(by='Amount', ascending =False).head(10)

better_graph = sns.barplot(data = state_vs_amount, x ='State', y='Amount')
better_graph.set_xticklabels(better_graph.get_xticklabels(), rotation=45)
sns.set(rc={"figure.figsize":(15, 6)})
```

```python
state_vs_orders = df.groupby(['State'], as_index =False)['Orders'].sum().sort_values(by='Orders', ascending =False).head(10)

better_graph = sns.barplot(data = state_vs_orders, x ='State', y='Orders')
better_graph.set_xticklabels(better_graph.get_xticklabels(), rotation=45)
sns.set(rc={"figure.figsize":(15, 6)})
```

FROM ABOVE 2 GRAPHS WE CAN SEE THAT HARYANA AND KERALA HAS SAME ORDERS, BUT HARYANA HAD HIGHER PURCHASING POWER THAN KERALA

```python
# Marital Status VS Amount
maritalStatusVSamount = df.groupby(['Marital_Status','Gender'], as_index =False)['Amount'].sum().sort_values('Amount', ascending =False)

sns.barplot(data =maritalStatusVSamount, x='Marital_Status', y='Amount', hue='Gender')
sns.set(rc={"figure.figsize":(7, 5)})
```

OCCUPATION

```python
plt.figure(figsize=(20, 5))
graph4 = sns.countplot(data= df, x='Occupation')

for bars in graph4.containers:
    graph4.bar_label(bars)
```

TO SEE WHICH OCCUPATION HAS HIGHEST PURCHASING POWER


```python
occupation_vs_amount = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)

plt.figure(figsize=(20, 5))
graph5 = sns.barplot(data=occupation_vs_amount, x='Occupation', y='Amount')

for bars in graph5.containers:
    graph4.bar_label(bars)
    
```

PRODUCT CATEGORY VS AMOUNT

```python
plt.figure(figsize=(20, 5))
graph6 = sns.countplot(data =df, x='Product_Category')
graph6.set_xticklabels(graph6.get_xticklabels(), rotation=45)

for bars in graph6.containers:
    graph6.bar_label(bars)
```

```python
product_vs_amount = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False).head(10)

graph7 = sns.barplot(data = product_vs_amount, x='Product_Category', y='Amount')
graph7.set_xticklabels(graph7.get_xticklabels(), rotation=45)
plt.figure(figsize=(20, 6))
```
